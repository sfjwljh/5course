```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import pandas as pd
# S=50.0
# sigma=0.385191
# rf=0.03
# q=0.0
# T=1.0
# X=52.0

def BS_call(S, sigma, rf, q, T, X):
   d1=(np.log(S/X)+(rf-q+sigma**2/2)*T)/(sigma*np.sqrt(T))
   d2=d1-sigma*np.sqrt(T)
   return S*np.exp(-q*T)*st.norm.cdf(d1)-X*np.exp(-rf*T)*st.norm.cdf(d2)
def BS_put(S, sigma, rf, q, T, X):
   d1=(np.log(S/X)+(rf-q+sigma**2/2)*T)/(sigma*np.sqrt(T))
   d2=d1-sigma*np.sqrt(T)
   return X*np.exp(-rf*T)*st.norm.cdf(d2)-S*np.exp(-q*T)*st.norm.cdf(d1)
#构建x、y数据
data=pd.read_excel('3d_read.xlsx',sheet_name=0)
S_ = data['S'].values
sigma_=data['sigma'].values
rf_=data['rf'].values
q_=data['q'].values
T_=data['T'].values
X_=data['X'].values

x=S_
y = T_
#将数据网格化处理
X, Y = np.meshgrid(x, y)
Z = BS_call(S=X,sigma=0.385191,rf=0.03,q=0.0,T=Y,X=52.0)
fig = plt.figure()
ax = plt.axes(projection='3d')
#50表示在z轴方向等高线的高度层级，binary颜色从白色变成黑色
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```
